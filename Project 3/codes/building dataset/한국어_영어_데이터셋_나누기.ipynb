```python
import numpy as np
from pandas import DataFrame
from sklearn.model_selection import train_test_split

#구어체
A1 = pd.read_excel("/content/drive/MyDrive/groom_project3/1_구어체(1).xlsx")
B1 = pd.read_excel("/content/drive/MyDrive/groom_project3/1_구어체(2).xlsx")
C1 = pd.read_excel("/content/drive/MyDrive/groom_project3/2_대화체.xlsx")
```

```python
A1_train, A1_test = train_test_split(A1, test_size=0.2, random_state=42)
A1_train, A1_val = train_test_split(A1_train, test_size=0.125, random_state=42)

B1_train, B1_test = train_test_split(B1, test_size=0.2, random_state=42)
B1_train, B1_val = train_test_split(B1_train, test_size=0.125, random_state=42)

C1_train, C1_test = train_test_split(C1, test_size=0.2, random_state=42)
C1_train, C1_val = train_test_split(C1_train, test_size=0.125, random_state=42)

print("구어체1 train, 구어체1 val, 구어체1 test", len(A1_train), len(A1_val), len(A1_test))
print("구어체2 train, 구어체2 val, 구어체2 test", len(B1_train), len(B1_val), len(B1_test))
print("대화체 train, 대화체2 val, 대화체2 test", len(C1_train), len(C1_val), len(C1_test))
print(" ")
print("구어체1", len(A1_train) + len(A1_val) + len(A1_test))
print("구어체2", len(B1_train) + len(B1_val) + len(B1_test))
print("대화체", len(C1_train) + len(C1_val) + len(C1_test))
print(" ")
print("train", len(A1_train) + len(B1_train) + len(C1_train))
print("val", len(A1_val) + len(B1_val) + len(C1_val))
print("test", len(A1_test) + len(B1_test) + len(C1_test))
```

```python
columns = ['원문', '번역문']

train_merge_df = pd.DataFrame()
val_merge_df = pd.DataFrame()
test_merge_df = pd.DataFrame()

A1_train_df = pd.DataFrame(A1_train, columns=columns)
B1_train_df = pd.DataFrame(B1_train, columns=columns)
C1_train_df = pd.DataFrame(C1_train, columns=columns)

A1_val_df = pd.DataFrame(A1_val, columns=columns)
B1_val_df = pd.DataFrame(B1_val, columns=columns)
C1_val_df = pd.DataFrame(C1_val, columns=columns)

A1_test_df = pd.DataFrame(A1_test, columns=columns)
B1_test_df = pd.DataFrame(B1_test, columns=columns)
C1_test_df = pd.DataFrame(C1_test, columns=columns)

train_merge_df = train_merge_df.append(A1_train_df, ignore_index=False)
train_merge_df = train_merge_df.append(B1_train_df, ignore_index=False)
train_merge_df = train_merge_df.append(C1_train_df, ignore_index=False)

val_merge_df = val_merge_df.append(A1_val_df, ignore_index=False)
val_merge_df = val_merge_df.append(B1_val_df, ignore_index=False)
val_merge_df = val_merge_df.append(C1_val_df, ignore_index=False)

test_merge_df = test_merge_df.append(A1_test_df, ignore_index=False)
test_merge_df = test_merge_df.append(B1_test_df, ignore_index=False)
test_merge_df = test_merge_df.append(C1_test_df, ignore_index=False)

train_merge_df.to_csv("구어체_train.csv", index=False, encoding = "utf-8-sig")

val_merge_df.to_csv("구어체_val.csv", index=False, encoding = "utf-8-sig")

test_merge_df.to_csv("구어체_test.csv", index=False, encoding = "utf-8-sig")
```

```python
news1 = pd.read_excel("/content/drive/MyDrive/groom_project3/3_문어체_뉴스(1)_200226.xlsx")
news2 = pd.read_excel("/content/drive/MyDrive/groom_project3/3_문어체_뉴스(2).xlsx")
news3 = pd.read_excel("/content/drive/MyDrive/groom_project3/3_문어체_뉴스(3).xlsx")
news4 = pd.read_excel("/content/drive/MyDrive/groom_project3/3_문어체_뉴스(4).xlsx")
culture = pd.read_excel("/content/drive/MyDrive/groom_project3/4_문어체_한국문화.xlsx")
law = pd.read_excel("/content/drive/MyDrive/groom_project3/5_문어체_조례.xlsx")
web = pd.read_excel("/content/drive/MyDrive/groom_project3/6_문어체_지자체웹사이트.xlsx")
```

```python
news1_cut = pd.DataFrame.sample(news1, n = 25000, replace=False, random_state = 42)
news1_train, news1_val = train_test_split(news1_cut, test_size=0.125, random_state=42)

news2_cut = pd.DataFrame.sample(news2, n = 25000, replace=False, random_state = 42)
news2_train, news2_val = train_test_split(news2_cut, test_size=0.125, random_state=42)

news3_cut = pd.DataFrame.sample(news3, n = 25000, replace=False, random_state = 42)
news3_train, news3_val = train_test_split(news3_cut, test_size=0.125, random_state=42)

news4_cut = pd.DataFrame.sample(news4, n = 25000, replace=False, random_state = 42)
news4_train, news4_val = train_test_split(news4_cut, test_size=0.125, random_state=42)

culture_cut = pd.DataFrame.sample(culture, n = 100000, replace=False, random_state = 42)
culture_train, culture_val = train_test_split(culture_cut, test_size=0.125, random_state=42)

law_cut = pd.DataFrame.sample(law, n = 100000, replace=False, random_state = 42)
law_train, law_val = train_test_split(law_cut, test_size=0.125, random_state=42)

web_cut = pd.DataFrame.sample(web, n = 100000, replace=False, random_state = 42)
web_train, web_val = train_test_split(web_cut, test_size=0.125, random_state=42)

print("뉴스1 train, 뉴스1 val", len(news1_train), len(news1_val))
print("뉴스2 train, 뉴스2 val", len(news2_train), len(news2_val))
print("뉴스3 train, 뉴스3 val", len(news3_train), len(news3_val))
print("뉴스4 train, 뉴스4 val", len(news4_train), len(news4_val))

print("문화 train, 문화 val", len(culture_train), len(culture_val))
print("법률 train, 법률 val", len(law_train), len(law_val))
print("웹사이트 train, 웹사이트 val", len(web_train), len(web_val))

print(" ")
print("train", len(news1_train) + len(news2_train) + len(news3_train) + len(news4_train) + len(culture_train) + len(law_train) + len(web_train))
print("val", len(news1_val) + len(news2_val) + len(news3_val) + len(news4_val) + len(culture_val) + len(law_val) + len(web_val))
```

```python
columns = ['원문', '번역문']

train_merge2_df = pd.DataFrame()
val_merge2_df = pd.DataFrame()

news1_train_df = pd.DataFrame(news1_train, columns=columns)
news2_train_df = pd.DataFrame(news2_train, columns=columns)
news3_train_df = pd.DataFrame(news3_train, columns=columns)
news4_train_df = pd.DataFrame(news4_train, columns=columns)
culture_train_df = pd.DataFrame(culture_train, columns=columns)
law_train_df = pd.DataFrame(law_train, columns=columns)
web_train_df = pd.DataFrame(web_train, columns=columns)

news1_val_df = pd.DataFrame(news1_val, columns=columns)
news2_val_df = pd.DataFrame(news2_val, columns=columns)
news3_val_df = pd.DataFrame(news3_val, columns=columns)
news4_val_df = pd.DataFrame(news4_val, columns=columns)
culture_val_df = pd.DataFrame(culture_val, columns=columns)
law_val_df = pd.DataFrame(law_val, columns=columns)
web_val_df = pd.DataFrame(web_val, columns=columns)

train_merge2_df = train_merge2_df.append(news1_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(news2_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(news3_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(news4_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(culture_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(law_train_df, ignore_index=False)
train_merge2_df = train_merge2_df.append(web_train_df, ignore_index=False)

val_merge2_df = val_merge2_df.append(news1_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(news2_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(news3_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(news4_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(culture_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(law_val_df, ignore_index=False)
val_merge2_df = val_merge2_df.append(web_val_df, ignore_index=False)


train_merge2_df.to_csv("문어체_train.csv", index=False, encoding = "utf-8-sig")

val_merge2_df.to_csv("문어체_val.csv", index=False, encoding = "utf-8-sig")
```
